```python
import scipy
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

# Chi Square

## Compare two distributions

```python
observed_values=scipy.array([18,21,16,7,15])
expected_values=scipy.array([22,19,44,8,16])

stats.chisquare(observed_values, f_exp=expected_values)
```

```python
plt.plot(observed_values, label = 'observed_values')
plt.plot(expected_values, label = 'expected_values')
plt.legend()
plt.show()
```

# Kolmogorov - Smirnov

```python
ab_test = pd.read_csv('ab_test.csv')
ab_test['a'].plot()
```

```python
## fit the distribution to your data and then check
loc, scale = stats.norm.fit(ab_test['a'])
n = stats.norm(loc=loc, scale=scale)
stats.kstest(ab_test['a'], n.cdf)
```
